```python
class DNode:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self,value):
        if not self.head:
            self.head = DNode(value)
        else:
            current = self.head
            while current.next:
                current = current.next
            new_node = DNode(value)
            current.next = new_node
            new_node.prev = current
            
    def __str__(self):
        out = "[%s" % self.head.data
        current = self.head.next
        while current:
            out += ", %s" % current.data
            current = current.next
        out += "]"
        return out
    
    def get_tail(self):
        current = self.head
        while current.next:
            current = current.next
        return current
    
    def rstr(self):
        current = self.get_tail()
        out = "[%s" % current.data
        current = current.prev
        while current:
            out += ", %s" % current.data
            current = current.prev
        out += "]"
        return out
    
    def __len__(self):
        counter = 0
        current = self.head
        while current:
            current = current.next
            counter += 1
        return counter
    
    def insert(self, index, value):
        # set error for negative index inputs
        if index < 0:
            raise IndexError("Cannot input negative index")
        new_node = DNode(value)
        if index == 0:
            temp = self.head
            self.head = new_node
            new_node.next = temp
        else:
            current = self.head
            counter = 0
            while current.next and index != counter:
                current = current.next
                counter += 1
            if index > counter:
                current.next = new_node
                new_node.prev = current
            elif index == counter:
                prev = current.prev
                prev.next = new_node
                new_node.prev = prev
                new_node.next = current
                current.prev = new_node
                
    def count(self,value):
        current = self.head
        counter = 0
        while current:
            if current.data == value:
                counter +=1
            current = current.next
        return print(value, "is in the list", counter,"times")
    
    def get_index(self, value):
        current = self.head
        ctr = 0
        i = 0
        while current:
            ctr += 1
            if current.data == value:
                i += 1
                break
            current = current.next
        if i == 1:
            print(value, " is found at index ", ctr)
        else:
            print(value, " not found!")
            
    def remove(self, value):
        if self.head.data == value:
            self.head = self.head.next
        else:
            current = self.head
            while current.next and current.data != value:
                prev = current
                current = current.next
            if current.data == value:
                prev.next = current.next
                prev.prev = current.prev
            else:
                raise ValueError("%s is not in the list" % value)
                
                
    def reverse(self):
        prev = self.head
        temp = self.head
        current = self.head.next
        prev.next = None
        prev.prev = None
        while current:
            temp = current.next
            current.next = prev
            prev.prev = current
            prev = current
            current = temp
        self.head = prev
        
                
        
        
           
        
        
        
        
```

```python
from functools import lru_cache

@lru_cache
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
```

```python
fib(100)
```

```python
my_cache = {}

def fib_d(n):
    val =my_cache.get(n)
    if val:
        return val
    if n < 2:
        return n
    my_cache[n-1] = fib_d(n-1)
    my_cache[n-2] = fib_d(n-2)
    return my_cache[n-1] + my_cache[n-2]
```
